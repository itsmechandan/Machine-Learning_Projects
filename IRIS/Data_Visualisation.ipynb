```python
import pandas as pd
import os
import kagglehub

# Download the dataset and get the path
path = kagglehub.dataset_download("uciml/iris")

# Construct the full path to the CSV file
# Replace 'Iris.csv' with the actual filename if it's different
csv_file_path = os.path.join(path, 'Iris.csv')

# Load the data into a pandas DataFrame
df = pd.read_csv(csv_file_path)

import seaborn as sns
import matplotlib.pyplot as plt

```

We can see some pattern in the Mean values of different parameters in each Species Type.
So Better we Data Visualise the Data Set first.


```python
# The first way we can plot things is using the .plot extension from Pandas dataframes
# We'll use this to make a scatterplot of the Iris features.
df.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# We can also use the seaborn library to make a similar plot
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=df, size=5)
```

```python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(df, hue="Species").map(plt.scatter, "SepalLengthCm", "SepalWidthCm").add_legend()
```

```python
sns.scatterplot(data = df, x= 'SepalLengthCm', y = "SepalWidthCm",hue = 'Species')
```

```python

# From the pairplot, we'll see that the Iris-setosa species is separataed from the other
# two across all feature combinations
sns.pairplot(df.drop("Id", axis=1), hue="Species", size=3)
```

To see how Logarithmic Transformation Works Just Created a new Data Frame for fun

```python
import numpy as np
normalised = df.copy()

# Apply log transformation to the 'petal_length' column
normalised['PetalLengthCm'] = np.log(normalised['PetalLengthCm'])
normalised['PetalWidthCm'] = np.log(normalised['PetalWidthCm'])
normalised['SepalLengthCm'] = np.log(normalised['SepalLengthCm'])
normalised['SepalWidthCm'] = np.log(normalised['SepalWidthCm'])

# Generate a pair plot of the new 'normalised' DataFrame
sns.pairplot(normalised, hue="Species")
```
